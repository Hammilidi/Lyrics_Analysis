# Importation des libriairies

```python
import requests
from bs4 import BeautifulSoup
from pprint import pprint
```

```python
def extract_lyrics(url):
    print("Fetching lyrics ...")
    r = requests.get(url)
    
    if r.status_code != 200:
        print("Page impossible à récupérer")
        return []

    soup = BeautifulSoup(r.content, 'html.parser')
    lyrics = soup.find("div", class_="lyrics")
    
    if not lyrics:
        return extract_lyrics(url)

    # Extract and print the lyrics as a string
    lyrics_text = '\n'.join(lyrics.stripped_strings)
    print(lyrics_text)
    
    return lyrics_text  # You can also return the lyrics as a string if needed

# Example usage:
# lyrics = extract_lyrics("https://example.com/lyrics")
```
